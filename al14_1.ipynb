```python
# 리스트 -> 리스트 : 컴프리헨션

# 계산에서 사용되는 계산 데이터셋(원본X)
values = [r for r in sales if isinstance(r["판매액"], int)]
len(values) # 결측치, 이상치를 제거한 18개가 나오는지 확인
```

```python
# 판매액 기준 최댓값, 최솟값 찾기

max(values, key = lambda x: x["판매액"]) # 람다
min(values, key = lambda x: x["판매액"])
```

```python
# Top-N 추출

sort_values = sorted(values, key = lambda x: x["판매액"], reverse=True)

rank_num = 1
for r in sort_values[:3]:
    print(f"{rank_num}등 {r}")
    rank_num +=1
```

```python
# 빈도수 계산

count = {} # 데이터를 저장하기 위한 빈 딕셔너리 생성
for r in sales:
    key_name = r["상품"]
    if key_name not in count:
        count[key_name] = 0
    count[key_name] = count[key_name] + 1

for k, v in count.items():
    print(f"{k}는 {v}건 입니다")
```

```python
# 그룹별 집계(Group By)

local_sum = {}

for r in sales:
    price = r["판매액"]
    # price "", "결측" 이면 
    if not isinstance(price, int):
        continue # 다음 반복으로 건너뛰기
    key_name = r["지역"]

    if key_name not in local_sum:
        local_sum[key_name] = 0
    local_sum[key_name] += price    

local_sum
```

```python
#  중복 제거

regions  = set()
for r in sales:
    regions.add(r["지역"])
    
print(regions)
```

```python
# 순위 계산(Ranking)

# sort_values

rank = 1
for r in sort_values:
    print(f"{rank}위 - {r}")
    rank += 1
```

```python
# 이동평균(Moving Average)

# 슬라이딩 윈도우에 사용될 리스트

prices = []

for r in sales:
    if isinstance(r["판매액"], int):
        prices.append(r["판매액"]) 
prices

window = 3

for i in range(len(prices) - (window - 1)):
    total = sum(prices[i: i+window])
    avg = round(total / window, 3)
    print(f"{i+1}~{i+window}일 평균 판매액 : {avg}원")
```

```python
# 누적평균(Running Average)

prices = []
for r in sales:
    if isinstance(r["판매액"], int):
        prices.append(r["판매액"])

# prices

prefix = [0]

for p in prices:
    prefix.append(prefix[-1] + p)

# prefix

prefix_avg = []

for i in range(1, len(prefix)):
    prefix_avg.append(prefix[i] / i)

prefix_avg
```

```python
# 이상치 탐지(Threshold) -> 처리는 문제에 맞는 방식으로 

for r in sales:
    if isinstance(r["판매액"], int):
        if r["판매액"] >= 2000000:
            print("이상치 발견", r)
```

```python
# 결측치 처리

# avg = sum(prices) / len(prices) # 평균

# 중앙값 # 정렬 먼저

prices = []

for r in sales:
    if isinstance(r["판매액"], (int, float)):
        prices.append(r["판매액"])


# 중앙값 계산
prices.sort()

if len(prices) % 2 == 1:
    median = prices[len(prices) // 2]
else:
    median = (
        prices[len(prices) // 2]
        + prices[len(prices) // 2 - 1]
    ) / 2


# 숫자가 아닌 판매액을 중앙값으로 변경
for r in sales:
    if not isinstance(r["판매액"], (int, float)):
        r["판매액"] = median


print("결측치->중앙값", sales)
```

```python
# 범주별 통계

result = {}

for r in sales:
    p_name = r["상품"]
    price = r["판매액"]
    total_cnt = r["수량"]
    if p_name not in result:
        result[p_name] = {"합계":0, "건수":0, "최대":price, "최소":price, "총수량":0}
    
    info = result[p_name]
    info["합계"] += price
    info["건수"] += 1
    info["총수량"] += total_cnt

# 최대/최소

if info["최대"] < price:
    info["최대"] = price

if info["최소"] > price:
    info["최소"] = price

print(result)
```
