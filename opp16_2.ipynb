```python
# 5. 객체의 상태 변경

class Product:
    def __init__(self, name, price, stock):
        self.name = name
        self.price = price
        self.stock = stock
    

    def total_price(self, count):
        return self.price * count

    def print_info(self):
        print(self.name)
        print(self.price)
        print(self.stock)

    def stock_in(self, count):
        self.stock = self.stock + count
    def stock_out(self, count):
        if self.stock >= count:
            self.stock = self.stock - count
        else:
            print("재고가 부족합니다.")
    def change_price(self, price):
        if price >= 0:
            self.price = price
        else:
            print("가격이 0원보다 작을 수 없습니다.")

products = [
    Product("키보드", 50000, 10),
    Product("마우스", 30000, 20),
    Product("모니터", 250000, 5),
    Product("노트북", 1200000, 3),
]


products[0].stock_in(5)
products[1].stock_out(3)
products[2].change_price(270000)

for p in products:
    p.print_info()
```

```python
# 6. 메서드를 이용한 기능 구현

class Product:
    def __init__(self, name, price, stock):
        self.name = name
        self.price = price
        self.stock = stock

    def order(self, count):
        if self.stock >= count:
            self.stock = self.stock - count
        else:
            print("재고 부족")

    def total_price(self, count):
        return self.price * count
    
    def is_sold_out(self):
        return self.stock == 0

products = [
    Product("키보드", 50000, 10),
    Product("마우스", 30000, 20),
    Product("모니터", 250000, 5),
    Product("노트북", 1200000, 3),
]


print(f"결제금액: {products[0].total_price(2)}원")
print(f"남은 재고: {products[0].order(2)}개")
```
