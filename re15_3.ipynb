```python
# 1. 문자열에서 숫자를 모두 찾아 출력하시오.

import re

text = "Python123AI456"
result = re.findall(r"[\d]+", text)
print(result)

# 출력
# ['123', '456']
```

```python
# 2. 문자열에서 소문자만 모두 출력하시오.

text = "PythonAI"
result = re.findall(r"[a-z]", text)
print(result)

# 출력
# ['y', 't', 'h', 'o', 'n']
```

```python
# 3. 문자열에 숫자가 포함되어 있는지 검사하시오.

text = "Python123"

if re.search(r"[\d]", text):
    print("숫자가 있습니다.")

# 출력
# 숫자가 있습니다.
```

```python
# 4. 문자열에서 전화번호를 찾아 출력하시오.

text = "문의 : 010-1234-5678"

result = re.search(r"\d{3}-\d{4}-\d{4}$", text)
print(result.group())

# 출력
# 010-1234-5678
```

```python
# 5. 문자열에서 이메일 찾아 출력하시오.

text = "문의 : python@test.com"

result = re.search(r"\w+@\w+\.\w+", text)
print(result.group())

# 출력
# python@test.com
```

```python
# 6. 문자열에서 숫자를 모두 *로 변경하시오.

text = "A10 B20 C30"
result = re.sub(r"\d", "*", text)
print(result)

# 출력
# A** B** C**
```

```python
# 7. 문자열에서 HTML태그("<"로 시작하고 ">"로 끝나는 문자)를 제거하시오.

html = "<h1>Python</h1>"
result = re.sub(r"<.+?>", "", html) # ?를 안 넣게 되면 처음 < 부터 끝 > 까지 하나로 인식함
print(result)

# 출력
# Python
```

```python
# 8. 문자열에서 날짜를 찾아 출력하시오.

text = """
오늘은 2026-07-07입니다.
시험은 2026-08-01입니다.
"""


result = re.findall(r"\d{4}-\d{2}-\d{2}", text)
print(result)

# 출력
# ['2026-07-07', '2026-08-01']
```
