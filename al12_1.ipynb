```python
# 시간복잡도

# 정확한 횟수를 계산하는 것이 아니라 데이터가 커질 때 얼마나 빨리 증가하는지를 보는 것
# 정확히 몇배가 늘어나는지는 알고리즘에 따라 다르다.
# 같은 O(n²) 알고리즘이라도 실제 비교 횟수와 실행 시간은 서로 다를 수 있다.
# O(n²)는 정확히 몇 배 증가하는지가 아니라, 데이터가 증가할수록 작업량이 대략 n²에 비례하여 증가함을 의미한다.
```

```python
# ===========================
# 선형 탐색 (O(n))
# ===========================


import random
import time

# 0~9999 사이의 숫자 1000개를 무작위로 생성
numbers = random.sample(range(10000), 1000)

# 실행 시간 측정 시작
start = time.time()
# start # 1970년 1월 1일 0시부터 지금까지 지난 시간을 초(Second) 단위로 나타낸 값

# 마지막 원소를 찾도록 설정(가장 오래 걸리는 상황(최악의 경우, Worst Case)을 테스트)
target = numbers[-1] # 마지막 원소를 찾으면 앞에서부터 끝까지 모두 확인해야 하니까

# 처음부터 끝까지 하나씩 비교
for n in numbers:
    if n == target:
        break

# 실행 시간 측정 종료
end = time.time()

print("선형 탐색 :", end - start) # 선형 탐색하는 데 걸린 시간
```

```python
# ===========================
# 버블 정렬 (O(n²))
# ===========================


# 새로운 무작위 데이터 생성
numbers = random.sample(range(10000), 1000) # 데이터가 늘어나면 늘어난 제곱만큼 시간이 더 소요됨

# 실행 시간 측정 시작
start = time.time()

# 버블 정렬 수행
for j in range(len(numbers) - 1):
    for i in range(len(numbers) - 1 - j):
        if numbers[i] > numbers[i + 1]:
            numbers[i], numbers[i + 1] = numbers[i + 1], numbers[i]

# 실행 시간 측정 종료
end = time.time()

print("버블 정렬 :", end - start)


# 선형 탐색은 데이터를 한 번만 훑기 때문에 매우 빠름
# 버블 정렬은 같은 데이터를 여러 번 반복해서 비교, 교환 - > 많은 연산 수행 
# -> 데이터가 많아질수록 실행 시간 차이가 크게 벌어짐
# 시간복잡도 O(n)과 O(n²)의 차이를 보여주는 대표적인 예
```

```python
# 공간복잡도
# 프로그램이 실행되는 동안 추가로 사용하는 메모리의 양

# 시간복잡도 vs 공간복잡도

# 시간복잡도 : 얼마나 오래 걸리는가?
# 공간복잡도 : 메모리를 얼마나 사용하는가?

# ※ 알고리즘과의 관계

좋은 알고리즘은 실행 시간이 짧고(Time), 메모리 사용량도 적은(Space) 알고리즘이다.
하지만 둘은 트레이드오프(Trade-off) 관계인 경우가 많다.
- 메모리를 더 사용하면 실행 시간을 줄일 수 있다.
- 메모리를 아끼면 실행 시간이 늘어날 수 있다.
```

```python
# 오른쪽 삭제 pop

from collections import deque

dq = deque([10, 20, 30])
x = dq.pop() # like 스택

print(x)
print(dq)

# 왼쪽 삭제 popleft

from collections import deque

dq = deque([10, 20, 30])
x = dq.popleft() # like 큐

print(x)
print(dq)
```

```python
from collections import deque

dq = deque([1, 2, 3, 4, 5])
dq.rotate(1) # 우측으로 한칸 이동(회전)

print(dq) # deque([5, 1, 2, 3, 4])


from collections import deque

dq = deque([1, 2, 3, 4, 5])
dq.rotate(-1) # 왼쪽으로 두 칸 회전 (반대방향)

print(dq)
```

```python
# 카드 게임(원카드, 우노, ..) 시뮬레이션

from collections import deque

cards = deque([1, 2, 3, 4, 5]) # 처음 받은 카드 5장

while len(cards) > 1:
    print("버림 :", cards.popleft()) # 1번 카드를 버림
    cards.rotate(-1) # 반대 방향으로 돌려야 하니까
    # cards.append(cards.popleft())
    print("남은 카드 :", cards)

print("마지막 카드 :", cards[0])
```
