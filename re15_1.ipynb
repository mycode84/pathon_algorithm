```python
import re

# ^(캐럿 caret) not
text = "a7bc123"
result = re.findall(r"[^0-9]", text) # 0~9문자가 아니면 반환되는 리스트에 추가
# ^ 캐럿문자가 대괄호 밖에 있을 경우는 0-9로 시작하는 문자임을 나타냄
print(result)
```

```python
import re

text = "cat ct cot cut coot"
result = re.findall(r"c.t", text) # "c_t" 이런 패턴의 문자(c로 시작, t로 끝나는 문자열길이 3) 찾기
print(result)
```

```python
import re

# findall함수에서는 ^패턴을 사용해도 의미있는 반환값을 얻을 수 없다(^패턴은 전체 문자열을 하나로 다루기 때문)
text = "Python Python python"
print(re.findall(r"^Py", text)) # ["Py"]

text = "python Python python"
print(re.findall(r"^Py", text)) # [] 

# ^010-(\d{4})-(\d{4})$
```

```python
# search 함수는 위치를 찾음
import re

text = "Python Python python"
print(re.search(r"^Py", text)) # <re.Match object; span=(0, 2), match='Py'> span은 Py문자의 위치 0에서 2-1에 위치
print(re.search(r"^py", text)) # None
```

```python
# match 함수를 사용하면 ^ 생략 가능 -> match 함수 자체가 시작점을 찾는 함수

print(re.search(r"^Py", text)) 
print(re.match(r"^Py", text))
print(re.match(r"Py", text))
print(re.match(r"Py", text).group())

# search 함수는 문자열 전체를 검색
# match 함수는 문자열 시작만 검색하기 때문에 아래 3가지는 같은 결과 
```

```python
# * 앞에 문자는 0번이상 여러번 반복될 수 있다. 최소 개수 0개
# + 앞에 문자는 1번이상 여러번 반복될 수 있다. 최소 개수 1개

import re

text = "Pthon"

# 시작이 Py 찾기
print(re.search(r"^Py", text)) # <re.Match object; span=(0, 2), match='Py'>
# 시작이 Pyyyyy.... 찾기
print(re.search(r"^Py*", text)) # <re.Match object; span=(0, 6), match='P'>
print(re.search(r"^Py+", text)) # None
```

```python
# match()함수 : 문자열 처음부터 패턴이 맞는지만 검사
# fullmatch()함수 : 문자열 전체가 패턴과 일치해야 함 

import re

texts = ["123", "", "abc", "12a"]

for t in texts:
    print(t, "->", bool(re.fullmatch(r"\d+", t))) # \d* : \d+ : 숫자가 1개 이상 있어야 한다.
```

```python
# 앞 문자가 0개 또는 1개 존재한다. "."은 무조건 1개 이상 
# "ab?c" : b가 없거나 1개 있거나 -> "abc"(O) 와 "ac"(O)는 매칭,  "abbc"(X) b가 2개라서 매칭 안됨

import re

words = [
    "ae",
    "abc",
    "abbcd",
    "abbbcde",
    "acde"
]

for word in words:
    print(re.match(r"ab?c", word))
# None
# <re.Match object; span=(0, 3), match='abc'>
# None
# None
# <re.Match object; span=(0, 2), match='ac'>

# span은 매칭된 문자열이 원본 문자열의 몇 번째 인덱스부터 몇 번째 인덱스까지인지를 나타냄
```

```python
import re

products = [
    "사과 3500원",
    "배 4200원",
    "수박 18000원"
]

price_list = []

for product in products:
    name = re.search(r"[가-힣]+", product)
    price = re.search(r"\d+", product)
    # print(price.group())
    price_list.append({name.group():price.group()})

print(price_list)
```

```python
import re

emails = [
    "python@네이버.com",
    "kim123@gmail.com",
    "abc@company.co.kr", # .co는 [A-Za-z0-9.-] 에 포함됨
    "hello",
    "@gmail.com"
]

pattern = r"^[A-Za-z0-9._%+-]+@[A-Za-z0-9.-]+\.[A-Za-z]{2,}$" # \. "." 이후에 두자 이상으로 끝남

for email in emails:
    if re.fullmatch(pattern, email):
        print(email, "정상")
    else:
        print(email, "오류")
```

```python
import re

phones = [
    "02-1234-5678",
    "010-1234-5678",
    "010-123-5678",
    "010-1234-11111"
]

# "숫자3시작-숫자4-숫자4끝나는" 찾기
for p in phones:
    result = re.search(r"^(\d{3})-(\d{4})-(\d{4})$", p)
    if result != None:
        print(result)
        print(result.group()) # group()함수 매칭된 결과물을 반환
```

```python
import re

urls = [
    "https://www.google.com",
    "http://naver.com",
    "https://python.org/docs",
    "google.com"
]

pattern = r"^https?://([\w-]+\.)+[\w-]+(/[^\s]*)?$"
# ^https? -> "http://" or "https://"로 시작
# ([\w-]+\.)+ -> 알파벳(\w) or "-" 가 1개 이상 / "\." 는 그냥 문자의 "."이 옴
# [\w-]+ -> 알파벳(\w) or "-" 가 1개 이상
# (/[^\s]*)?$ -> (/[^\s]*) 있어도 되고 없어도 됨. "/" 에 공백이 아닌 모든 것이 0번 이상 반복

for url in urls:
    if re.fullmatch(pattern, url):
        print(url, "정상")
    else:
        print(url, "오류")
```

```python
import re

nums = [
    "901010-1234567",
    "000101-4234567",
    "9010101234567",
    "90101-1234567"
]

pattern = r"^(\d{6})-([1-4]\d{6})$"

for n in nums:
    if re.fullmatch(pattern, n):
        print(n, "형식 정상")
    else:
        print(n, "오류")
```
