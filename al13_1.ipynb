```python
scores = [10, 20, 3, 40, 5, 60, 8, 55, 44]

k = 3

# window 시작 포인트(인덱스 0)의 합을 초기값으로 설정
window = sum(scores[0:k]) # scores[0:3]

total = [window]
avgs = [window / k]

# 최댓값과 최솟값의 초기값 설정
max_val = {"구간": "0~2", "합":window}
min_val = {"구간": "0~2", "합":window} 

# print(window)

# k부터 시작하는 이유 -> 첫 번째 구간인 0~2는 이미 계산했기 때문
for i in range(k, len(scores)): # 3칸 이동 # 원도우 크기 부터 1칸식 이동하며 윈도우 안의 합을 연산
    window = window + scores[i] - scores[i-k]

    # print(window)

    total.append(window)
    avgs.append(window / k) # 사이즈로 나눠서 평균 구하기

    section = f"{i - (k-1)} ~ {i}" 
    # 현재 끝 인덱스 i를 포함해서 총 k개의 요소를 세려면, 뒤로 k-1칸 이동해야 시작 인덱스가 되기 때문
    # 핵심: i 자신도 윈도우의 1개 요소로 포함되기 때문에 k칸이 아니라 k-1칸만 뒤로 간다는 것

    if window > max_val["합"]:
        max_val = {"구간":section, "합":window}

    if window < min_val["합"]:
        min_val = {"구간":section, "합":window}

print(total) # 누적합
print(avgs) # 누적평균
print(max_val) # 최댓값
print(min_val) # 최솟값
```

```python
fruit = {
    "color": "Red",
    "size": "Large" # Small
}

fruit_tree = {
    "question": ("color", "Red"),

    "yes": {
        "question": ("size", "Large"),
        "yes": "사과 (Apple)",
        "no": "체리 (Cherry)"
    },

    "no": {
        "question": ("color", "Purple"),
        "yes": "포도 (Grape)",
        "no": "모르는 과일"
    }
}

node = fruit_tree

# 트리를 탐색하는 반복문
while isinstance(node, dict): # node객체가 dict 이라면 True 아니면 False

    # dict이라면 "question"키값을 찾아 질문은 key, 정답은 value에 저장 
    key, value = node["question"] 
   
    # 입력데이터의 key값과 비교하여 정답이면 "yes"키값을 node에 대입, 아니면 "no"키값을 node에 대입
    # 이렇게 대입된 node값이 dict 타입이면 다시 반복문 진행... 
    # 이렇게 node에 dict타입이 아닌 타입(위 트리에선 문자열)이 올때 까지 반복문은 계속
    if fruit[key] == value:
        node = node["yes"]
    else:
        node = node["no"]

# 반복문을 빠져 나오면 마지막 node값을 출력
print(node)
```
