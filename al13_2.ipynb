```python
# 실습 과제 1. 누적합(Prefix Sum)

steps = [5231, 6120, 4890, 7005, 6500, 7200, 6800]
prefix = [0]

for s in steps:
    prefix.append(prefix[-1] + s)

for i in range(1,len(prefix)):
    print(f"{i}일 누적 : {prefix[i]}")
```

```python
# 실습 과제 2. 누적합(Prefix Sum) - 1번 응용

steps = [
    {"2026-01-01": 5231},
    {"2026-01-02": 6210},
    {"2026-01-03": 4890},
    {"2026-01-04": 7005},
    {"2026-01-05": 6500},
    {"2026-01-06": 7200},
    {"2026-01-07": 6800}
]

prefix = [0]

for data in steps: # 일단 step부터 읽기
    for date, value in data.items(): # 딕셔너리 안 키와 값을 동시에 읽을 items를 사용하기
        prefix.append(prefix[-1] + value) # date, value로 구분했으니 prefix에 넣어주기
        print(f"{date} 누적 : {prefix[-1]}") # 최신 누적한 마지막 값부터 출력하기 위해
```

```python
# 실습 과제 3. 슬라이딩 윈도우

sales = [12, 15, 18, 20, 17, 25, 22, 19]

window = sales[0] + sales[1] + sales[2]

total = [window]
max_val = window



for i in range(3, len(sales)): 
    window = window + sales[i] - sales[i-3]
    total.append(window)
    print(sales[i-3], "+", sales[i-2], "+", sales[i-1], "=", window)

    if window > max_val:
        max_val = window

print()
print("최대 매출 합 :", max_val)
```
