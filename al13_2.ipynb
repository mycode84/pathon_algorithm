```python
# мӢӨмҠө кіјм ң 1. лҲ„м Ғн•©(Prefix Sum)

steps = [5231, 6120, 4890, 7005, 6500, 7200, 6800]
prefix = [0]

for s in steps:
    prefix.append(prefix[-1] + s)

for i in range(1,len(prefix)):
    print(f"{i}мқј лҲ„м Ғ : {prefix[i]}")
```

```python
# мӢӨмҠө кіјм ң 2. лҲ„м Ғн•©(Prefix Sum) - 1лІҲ мқ‘мҡ©

steps = [
    {"2026-01-01": 5231},
    {"2026-01-02": 6210},
    {"2026-01-03": 4890},
    {"2026-01-04": 7005},
    {"2026-01-05": 6500},
    {"2026-01-06": 7200},
    {"2026-01-07": 6800}
]

prefix = [0]

for data in steps: # мқјлӢЁ stepл¶Җн„° мқҪкё°
    for date, value in data.items(): # л”•м…”л„ҲлҰ¬ м•Ҳ нӮӨмҷҖ к°’мқ„ лҸҷмӢңм—җ мқҪмқ„ itemsлҘј мӮ¬мҡ©н•ҳкё°
        prefix.append(prefix[-1] + value) # date, valueлЎң кө¬л¶„н–ҲмңјлӢҲ prefixм—җ л„Јм–ҙмЈјкё°
        print(f"{date} лҲ„м Ғ : {prefix[-1]}") # мөңмӢ  лҲ„м Ғн•ң л§Ҳм§Җл§ү к°’л¶Җн„° м¶ңл Ҙн•ҳкё° мң„н•ҙ
```
